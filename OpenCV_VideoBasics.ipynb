```python
import cv2
```

```python
cap = cv2.VideoCapture("C:\\Users\\SHYAM\\Downloads\\Video_Car.mp4")
```

```python
#Converting BGR to Gray scale
while (cap.isOpened()):
    ret, frame = cap.read()
    
    # convert BGR format to Grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('original',frame)
    cv2.imshow('Gray',gray_frame)
    
    k = cv2.waitKey(10) & 0xFF
    if k == ord('q') or k == 27:
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
#Capturing a frame and displaying frame count, height, width and frame per sec
import matplotlib.pyplot as plt
cap = cv2.VideoCapture("C:\\Users\\SHYAM\\Downloads\\Video_Car.mp4")

# Capture a Video Frame
ret, frame = cap.read()

frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

# imshow() - To display an Image
plt.imshow(frame)

print("Display Frame Count:",cap.get(cv2.CAP_PROP_FRAME_COUNT))
print("Display Frame Height:",cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print("Display Frame Width:",cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print("Display Frame FPS:",cap.get(cv2.CAP_PROP_FPS))

cap.release()
```

```python
#Increasing the video resolution and then adding average blur, median blur and gaussian blur

cap = cv2.VideoCapture("C:\\Users\\SHYAM\\Downloads\\Video_Car.mp4")
while (cap.isOpened()):
    ret, frame = cap.read()
    # increase the video resolution
    # fx = scale factor along horizontal axis
    # fy = scale factor along vertical axis
    frame = cv2.resize(frame, (840,680), fx = 200, fy = 300, 
                       interpolation = cv2.INTER_NEAREST)
    cv2.imshow('Original', frame)
    
    # Average Blur
    img_blur = cv2.blur(frame, (5,5))
    cv2.imshow('Blur', img_blur)

    # Median Blur
    m_blur = cv2.medianBlur(frame, 7)
    cv2.imshow('Median Blur', m_blur)

    # Gaussian Blur, sigmaX(stdDev)= 1
    g_blur = cv2.GaussianBlur(frame, (7,7), 1)
    cv2.imshow('Gaussian Blur', g_blur)
    
    k = cv2.waitKey(10) & 0xFF
    if k == ord('q') or k == 27:
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
#
cap = cv2.VideoCapture("C:\\Users\\SHYAM\\Downloads\\Video_Car.mp4")
while (cap.isOpened()):
    ret, frame = cap.read()
    frame = cv2.resize(frame, (840,680), fx = 200, fy = 300, 
                       interpolation = cv2.INTER_NEAREST)
    cv2.imshow('Original', frame)
    
    # convert BGR format to Grayscale
    img_g = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    edges_low = cv2.Canny(img_g, 50, 150)
    edges_high = cv2.Canny(img_g, 150, 220)
 
    cv2.imshow('original',frame)
    cv2.imshow('Edges',edges_low)
    cv2.imshow('Edges High',edges_high)
    
    k = cv2.waitKey(60) & 0xFF
    if k == ord('q') or k == 27:
        break
    
cap.release()
cv2.destroyAllWindows()
```
